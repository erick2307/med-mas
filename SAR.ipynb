```python
# make a barplot of four categories of data with error bars

import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C', 'IRCH']

# set width of bar
barWidth = 0.25

# set height of bar
bars1 = [210, 206, 204, 223]
mins = [191, 184, 184]
maxs = [223, 223, 223]

# set error bars
yer1 = [5.19, 4.38, 4.94, 0]

# set position of bar on X axis
r1 = np.arange(len(bars1))

cases_color = [95/255, 101/255, 154/255]
irch_color = [129/255, 186/255, 187/255]
colors = [cases_color, cases_color, cases_color, irch_color]

# make the plot
plt.bar(r1, bars1, width=barWidth, color=colors, yerr=yer1, capsize=5, zorder=10)
plt.scatter(r1[:-1], mins, color='black', marker='d', zorder=11)
for i in range(len(mins)):
    plt.annotate('Min', xy=(r1[i]+0.05, mins[i]-3), color='black', zorder=15)

plt.scatter(r1[:-1], maxs, color='black', marker='d', zorder=11)
for i in range(len(maxs)):
    plt.annotate('Max', xy=(r1[i]+0.05, maxs[i]-3), color='black', zorder=15)

plt.hlines(223, 0, 3, color='black', linestyle='-.', linewidth=0.2, zorder=1)

# add xticks on the middle of the group bars
plt.xlabel('Case', fontweight='bold')
plt.ylabel('Average transported patients by case', fontweight='bold')
# plt.xticks([r + barWidth for r in range(len(bars1))], ['A', 'B', 'C', 'IRCH'])
plt.xticks(r1, labels)

# create legend & show graphic
plt.grid(alpha=0.3, linestyle='--', zorder=0)
plt.savefig('barplot.png', dpi=600)
plt.show()
```
